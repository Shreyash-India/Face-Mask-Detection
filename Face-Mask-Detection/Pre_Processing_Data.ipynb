```python
import numpy as np
import os, cv2

root = r"C:\Users\KIIT\Tensorflow\Face_mask_detection\dataset"

# cateogries {1: with_mask, 2: without_mask}
categories = os.listdir(root)
# Labels
labels = np.arange(2)
```

Organizing Data

```python
dim = 100        # Dimmension (100 X100)
lablesCateogry = {'with_mask': 0, 'without_mask' : 1}

Data, Labels = [], []

for category in categories:
    root_sub_folder = os.path.join(root, category)
    images = os.listdir(root_sub_folder)
    
    for image in images:
        img_path = os.path.join(root_sub_folder, image)
        img = cv2.imread(img_path)
        
        # 1: RGB -> Gray scale
        # 2: Resize to (100 X 100) Dimmension
        grayMode_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        resized_img = cv2.resize(grayMode_img, (dim, dim))
        
        Data.append(resized_img)
        Labels.append(lablesCateogry[category])
```

```python
print('Shape of Data', np.array(Data).shape)
print('Shape of Labels', np.array(Labels).shape)
```

Scaling Dataset

```python
# data [List -> np.array]
Data = np.array(Data)/255.0
Data = np.reshape(Data, (Data.shape[0], dim, dim, 1))

# Labels [List -> np.array]
Labels = np.array(Labels)
```

```python
print('Shape of Data', np.array(Data).shape)
print('Type of Data', type(Data))

print('Shape of Labels', np.array(Labels).shape)
print('Type of Labels', type(Labels))
```

```python
from keras.utils import np_utils
# Lables[uni-Dimmensional np.array -> bi-Dimmensional categorical numpy array(dtype = float32)]
Labels = np_utils.to_categorical(Labels)
Labels
```

```python
# Categorical Labels
print('Shape of Labels', np.array(Labels).shape)
print('Type of Labels', type(Labels))
```

Save the Data and Labels as numpy array

```python
# save DATA & LABELS 
np.save('Data', Data)
np.save('Labels', Labels)
```
